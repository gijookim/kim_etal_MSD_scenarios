# Figure 3: Regression Tree Results

```python
df_table_fds =  pd.read_csv(file_path + 'processed_data/RT_Results/RT_reg_fb_2050_g1.csv', index_col=[0])
df_table_eb  =  pd.read_csv(file_path + 'processed_data/RT_Results/RT_reg_reb_2050_g1.csv', index_col=[0])
df_table_pws =  pd.read_csv(file_path + 'processed_data/RT_Results/RT_reg_pws_2050.csv', index_col=[0])

df_table_fds  = df_table_fds.drop(columns=['%IncMSE','metric'])
df_table_eb   = df_table_eb.drop(columns=['%IncMSE','metric'])
df_table_pws  = df_table_pws.drop(columns=['%IncMSE','metric'])
```

```python
df_table_fds.columns = ['IncNodePurity','varnames','region']
df_table_eb.columns = ['IncNodePurity','varnames','region']
df_table_pws.columns = ['IncNodePurity','varnames','region']
```

```python
#Normalize the variable importances across basins
for rg in df_table_fds.region.unique()[:]: #loop through the basins
     df_filt = df_table_fds[df_table_fds.region==rg] #select basin 
     sum_val_fds = df_filt['IncNodePurity'].sum() #sum up the raw IncNodePurity variable importance values
     df_table_fds.loc[df_table_fds['region']==rg,'norm_IncNodePurity'] = df_table_fds['IncNodePurity']/sum_val_fds #normalize
        
#pivot the dataframe for heatmap plotting
df_pivot_fds = pd.pivot_table(data=df_table_fds,values='norm_IncNodePurity',index=['region'],columns=['varnames'])
df_pivot_fds = df_pivot_fds.filter(['soc','icd','co2','ag','fd','bee','ed','res','gcm']) #reorder columns (x-axis) of heatmap
```

```python
#Normalize the variable importances across basins
for rg in df_table_eb.region.unique()[:]: #loop through the basins
     df_filt = df_table_eb[df_table_eb.region==rg] #select basin 
     sum_val_eb = df_filt['IncNodePurity'].sum() #sum up the raw IncNodePurity variable importance values
     df_table_eb.loc[df_table_eb['region']==rg,'norm_IncNodePurity'] = df_table_eb['IncNodePurity']/sum_val_eb #normalize
        
#pivot the dataframe for heatmap plotting
df_pivot_eb = pd.pivot_table(data=df_table_eb,values='norm_IncNodePurity',index=['region'],columns=['varnames'])
df_pivot_eb = df_pivot_eb.filter(['soc','icd','co2','ag','fd','bee','ed','res','gcm']) #reorder columns (x-axis) of heatmap
```

```python
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib.colors import LogNorm

class MidPointLogNorm(LogNorm):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        LogNorm.__init__(self,vmin=vmin, vmax=vmax, clip=clip)
        self.midpoint=midpoint
    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        x, y = [np.log(self.vmin), np.log(self.midpoint), np.log(self.vmax)], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(np.log(value), x, y))
```
